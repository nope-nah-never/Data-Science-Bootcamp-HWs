Numpy Questions

```python
import numpy as np
```

Q1

```python
A = np.array([1, 5, 7, 9, 10, 3, 2, 5, 8, 9])
B = np.array([5, 8, 2, 8,  1, 3, 10, 2, 3, 6])

v_stacked = np.vstack([A, B])
h_stacked = np.hstack([A, B])

print(v_stacked)
print(h_stacked)
```

Q2

```python
common = np.intersect1d(A, B)
print(common)
```

Q3

```python
mask = (A >= 5) & (A <= 10)
vals_in_range = A[mask]
print(vals_in_range)
```

Q4

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
iris_mask = (iris_2d[:, 2] > 1.5) & (iris_2d[:, 0] < 5.0)
iris_filtered = iris_2d[iris_mask]
print(iris_filtered)
```
